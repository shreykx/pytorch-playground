```python
import torch
import matplotlib.pyplot as plt
import numpy as np
```

```python
lamfs = lambda x : x**2+20
```

```python

xs = np.arange(-10, 10, 0.75)
xs
```

```python
ys = lamfs(xs)
ys
```

```python
plt.plot(xs, ys)
```

```python
plt.scatter(xs, ys)
```

```python
plt.bar(xs, ys)
```

```python
labels = np.array(["dogs", "cats", "hippos"])
wedge = np.array([0.2, 0, 0])
plt.pie(torch.tensor([10, 60, 30]), labels=labels, startangle=90, explode=wedge) # default startangle is 360*x 
```

`explode will seperate the given param from the centre by the amount specified`
